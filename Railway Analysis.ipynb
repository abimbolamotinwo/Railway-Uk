```python
railway['Date of Purchase'] = pd.to_datetime(railway['Date of Purchase'])
railway['Date of Journey'] = pd.to_datetime(railway['Date of Journey'])
```

```python
railway.info()
```

```python
railway['Time of Purchase'] = pd.to_datetime(railway['Time of Purchase'], format='%H:%M:%S')
railway['Departure Time'] = pd.to_datetime(railway['Departure Time'], format='%H:%M:%S')
railway['Arrival Time'] = pd.to_datetime(railway['Arrival Time'], format='%H:%M:%S')
railway['Actual Arrival Time'] = pd.to_datetime(railway['Actual Arrival Time'], format='%H:%M:%S')
```

#### Question 1: What is the most popular routes

```python
routecounts=railway.groupby(['Departure Station', 'Arrival Destination']).size().reset_index(name='Route Count')
routecounts
```

```python
popular_routes=routecounts.sort_values('Route Count', ascending=False)
popular_routes.head(10)
```

#### Question 2: What are the peak travel times?

```python
peak_hours=railway['Departure Time'].dt.hour.value_counts().head()
peak_days=railway['Date of Journey'].dt.day_name().value_counts().head()
print("Peak top 5 travel hour:\n", peak_hours)

print("Peak top 5 travel day:\n", peak_days)
```

#### Question 3: How does revenue vary by ticket types and classes?

```python
revenue_by_ticket=railway.groupby(['Ticket Class', 'Ticket Type'])['Price'].sum().reset_index(name='Total Revenue')
revenue_by_ticket
```

#### Question 4: What is the on-time performance? What are the main contributing factors?

```python
ontime_performance=round(railway['Journey Status'].value_counts(normalize=True)*100, 2)
print('ontime_performance in % is', ontime_performance[0])
```

```python
#Filter delayed journeys and analyze reasons for delay
delay=railway[railway['Journey Status']=='Delayed']
delay
```

```python
# main contributing factors to delay
delayed_reasons=delay['Reason for Delay'].value_counts()
delayed_reasons
```
